```python
plt.scatter(df['time_study'],df['Marks'])
plt.title('Scatter Plot of timestudy vs marks')
plt.xlabel('time_study(hours)')
plt.ylabel('marks')
plt.grid(True)
plt.show()
```

```python
df_grouped = df.groupby('number_courses')['Marks'].mean().reset_index()
plt.bar(df_grouped['number_courses'], df_grouped['Marks'],color='skyblue')
plt.title('Bar plot of number of courses vs average marks')
plt.xlabel('number of courses')
plt.ylabel('average marks')
plt.grid(True)
plt.show()

```

```python
x = df[["number_courses",'time_study']]
y = df['Marks']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtrain
```

```python
ytrain
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
y_pred=model.predict(xtest)
```
